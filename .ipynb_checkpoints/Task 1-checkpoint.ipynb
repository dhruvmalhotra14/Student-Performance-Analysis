```python
import pandas as pd

df = pd.read_csv("student-mat.csv", sep=";")

print("Dataset Shape:", df.shape)

df.head()
```

```python
print(df.isnull().sum())
```

```python
print("Duplicate Rows:", df.duplicated().sum())
```

```python
avg_grade = df["G3"].mean()
print("Average Final Grade:", avg_grade)
```

```python
high_score = df[df["G3"] > 15]
print("Students scoring above 15:", len(high_score))
```

```python
correlation = df["studytime"].corr(df["G3"])
print("Correlation between Study Time and Final Grade:", correlation)
```

```python
gender_avg = df.groupby("sex")["G3"].mean()
print(gender_avg)
```

```python
import matplotlib.pyplot as plt

plt.hist(df["G3"], bins=10)
plt.title("Distribution of Final Grades")
plt.xlabel("Grade")
plt.ylabel("Number of Students")
plt.show()
```

```python
plt.scatter(df["studytime"], df["G3"])
plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade")
plt.show()
```

```python
gender_avg.plot(kind="bar")
plt.title("Average Score by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Grade")
plt.show()
```

# Conclusion

- The dataset contains 395 students and 33 attributes.
- No major data quality issues were found after checking for missing values and duplicates.
- The average final grade (G3) was calculated.
- A number of students scored above 15 in the final grade.
- Study time shows a relationship with academic performance.
- Average grades were compared between male and female students.
- Visualizations such as histogram, scatter plot, and bar chart helped understand grade distribution and performance trends.

This analysis provides useful insights into student performance and the factors that may influence academic results.
